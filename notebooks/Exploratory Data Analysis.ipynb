```python
fig, ax = plt.subplots( 5, 5, figsize = ( 10, 10 ) )
fig.tight_layout()
i = 0
for emotion in os.listdir( IMG_PATH ):
    j = 0
    images_list = os.listdir( os.path.join( IMG_PATH, emotion ) )
    num_images = len( images_list ) 
    for _ in range( num_images ):
        image = images_list[ random.randint( 0, num_images ) ]
        ax[i][j].imshow( cv2.imread( os.path.join( IMG_PATH, emotion, image ) ) )
        ax[i][j].set_title( emotion )
        j += 1
        if( j == 5 ):
            break
    i += 1
    if( i == 5 ):
        break
```
